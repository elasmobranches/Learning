```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os

# Data 경로 설정
DATA_DIR = "data"

# 경고 무시
warnings.filterwarnings("ignore")

SEED = 123
```

## 데이터 로드


```python
# train 데이터셋 로드 (train.csv)
train = pd.read_csv(os.path.join(DATA_DIR, "train.csv"))

# test 데이터셋 로드 (test.csv)
test = pd.read_csv(os.path.join(DATA_DIR, "test.csv"))
```

```python
X = train_data.drop("SalePrice", axis=1)
Y = train_data["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(
    X, Y, test_size=0.2, random_state=SEED
)
```

```python
x_test = test_data.drop("SalePrice", axis=1)
```

### CustomDataset 생성


```python
import torch
from torch.utils.data import Dataset
from sklearn.preprocessing import StandardScaler


class CustomDataset(Dataset):
    def __init__(self, x, y=None, normalize=True):
        super(CustomDataset, self).__init__()
        self.x = x
        self.y = y

        # 데이터 표준화
        if normalize:
            scaler = StandardScaler()
            self.x = pd.DataFrame(scaler.fit_transform(self.x))

        # 텐서 변환
        self.x = torch.tensor(self.x.values).float()

        if y is not None:
            self.y = torch.tensor(self.y.values).float()

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        x = self.x[idx]
        if self.y is not None:
            y = self.y[idx]
            return x, y
        else:
            return x
```

### Dataset 생성


```python
train_dataset = CustomDataset(x_train, y_train, normalize=True)
valid_dataset = CustomDataset(x_valid, y_valid, normalize=True)
test_dataset = CustomDataset(x_test, normalize=True)
```

### Data Loader 생성


```python
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
```

### GPU 설정


```python
if torch.backends.mps.is_built():
    # mac os mps 지원 체크
    device = torch.device("mps" if torch.backends.mps.is_built() else "cpu")
else:
    # cuda 사용 가능한지 체크
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

### 모델 정의


```python
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self, num_features):
        super(Net, self).__init__()
        # self.fc1 = nn.Linear(num_features, 32)
        # self.fc2 = nn.Linear(32, 8)
        self.fc = nn.Sequential(
            nn.Linear(num_features, 1024),
            nn.ReLU(),
            nn.Linear(1024, 256),
            nn.ReLU(),
            nn.Linear(256, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1),
        )
        # 마지막 출력층의 Neuron은 1개로 설정
        # self.output = nn.Linear(8, 1)

    def forward(self, x):
        # x = F.relu(self.fc1(x))
        # x = F.relu(self.fc2(x))
        # x = self.output(x)
        x = self.fc(x)
        return x
```

```python
# 컬럼수 정의
NUM_FEATURES = len(x_train.columns)
print(NUM_FEATURES)
```

### 모델 생성


```python
# 모델 생성
model = Net(NUM_FEATURES)
# 모델을 device 에 올립니다. (cuda:0 혹은 cpu)
model.to(device)
model
```

### 손실함수와 옵티마이저 정의


```python
# Mean Squared Error(MSE) 오차 정의
loss_fn = nn.MSELoss()
```

```python
def RMSE(y_hat, y):
    return torch.sqrt(torch.mean((y_hat - y) ** 2))


loss_fn = RMSE
```

```python
# 옵티마이저 설정: model.paramters()와 learning_rate 설정
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

### 학습(Training)


```python
def train(model, train_loader, optimizer, loss_fn, device):
    model.train()
    final_loss = 0
    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        # gradient 초기화
        optimizer.zero_grad()
        # forward pass
        y_hat = model(x)
        # loss 계산
        loss = loss_fn(y_hat, y.unsqueeze(1))
        # backward pass
        loss.backward()
        # weight update
        optimizer.step()
        final_loss += loss.item()
    final_loss /= len(train_loader)
    return final_loss


def evaluate(model, test_loader, loss_fn, device):
    # 모델을 평가 모드로 설정
    model.eval()
    final_loss = 0
    with torch.no_grad():

        for x, y in test_loader:
            # 데이터를 device에 올립니다.
            x, y = x.to(device), y.to(device)
            # forward pass
            y_hat = model(x)
            # loss 계산
            loss = loss_fn(y_hat, y.unsqueeze(1))
            # loss 누적
            final_loss += loss.item()
    final_loss /= len(test_loader)
    return final_loss
```

```python
# 최대 반복 횟수 정의
num_epoch = 2000

# loss 기록하기 위한 list 정의
losses = []
val_losses = []

MIN_LOSS = 9999999999.0

STATE_DICT_PATH = "house-model.pth"

for epoch in range(num_epoch):
    loss = train(model, train_loader, optimizer, loss_fn, device)
    val_loss = evaluate(model, valid_loader, loss_fn, device)
    losses.append(loss)
    val_losses.append(val_loss)
    if val_loss < MIN_LOSS:
        print(
            f"{epoch:05d} [INFO] Model saved... {MIN_LOSS:.5f} ===> {val_loss:.5f}")
        MIN_LOSS = val_loss
        torch.save(model.state_dict(), STATE_DICT_PATH)
# state dict는 모든 w의 가중치를 기록한다
# torch.save를 통해서 최저점의 가중치를 load 가능하다
    if epoch % 20 == 0:
        print(
            "{0:05d} loss = {1:.5f}, val_loss = {2:.5f}".format(
                epoch, loss, val_loss)
        )
```

```python
# 모델 불러오기
model.load_state_dict(torch.load(STATE_DICT_PATH))
```

### 예측 코드


```python
# 결과입력
your_answer = []

# evaluation mode 로 변경
model.eval()

with torch.no_grad():
    for x in test_loader:
        x = x.to(device)
        y_hat = model(x).cpu().detach().numpy().item()
        your_answer.append(y_hat)
```
