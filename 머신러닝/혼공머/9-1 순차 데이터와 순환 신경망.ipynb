### 순차 데이터는 텍스트나 시계열 데이터처럼 순서에 의미가 있는 데이터를 말한다 RNN(Recurrent Neural Network)
텍스트 데이터는 지금까지 보았던 데이터들과는 다르게 단어의 순서가 중요하다
따라서 순차 데이터를 다룰 때는 이전에 입력한 데이터를 기억하는 기능이 필요하다
완전 연결 신경망이나 합성곱 신경망은 이런 기억 장치가 없다
하나의 샘플을 사용하여 정방향 계산을 수행하고 나면 그 샘플을 버려지고 다음 샘플을 처리할 때 재사용하지 않는다
이렇게 입력 데이터의 흐름이 앞으로만 전달되는 신경망을 *피드포워드 신경망*이라고 한다
다음 샘플을 위해서 이전 데이터가 신경망 층에 순환되는 신경망을 *순환 신경망*이라고 한다
### 순환 신경망은 일반적인 완전 연결 신경망과 비슷하나 완전 연결 신경망에 이전 데이터의 처리 흐름을 순환하는 고리를 추가한다
뉴런의 출력이 다시 자기 자신으로 전달된다
샘플을 처리하는 한 단계를 *타임 스텝*이라고 말한다
순환 신경망은 이전 타임스텝의 샘플을 기억하지만 타임스텝이 오래될수록 순환되는 정보는 희미해진다
순환 신경망은 특별히 층을 *셀*이라고 부른다. 또 셀의 출력을 *은닉 상태*라고 부른다

일반적으로 은닉층의 활성화 함수로는 tanh가 많이 사용된다
순환 신경망도 활성화 함수가 당연하게 필요하다
