```python
g = sns.FacetGrid(mydata, hue='gender', height = 6)
g.map(sns.histplot, 'age', alpha=0.5)
plt.subplots_adjust(top=0.9)
g.figure.suptitle('Age Distribution by gender')
g.add_legend()
plt.show()
```

```python
mydata[['gender', 'age']].groupby('gender').mean()
```

```python
sns.countplot(x=mydata['married'] )
plt.title("Married or not-married distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(mydata['married'].value_counts(), labels=['Not-married','Married'], autopct='%1.1f%%', explode=(0, 0.1), colors=['#ff9999','#66b3ff'])
plt.title('Married or not-married distribution')
plt.show()
```

```python
# 성별에 따른 결혼 여부
mydata.groupby(['gender'])['married'].value_counts()
```

```python
# object 타입의 문자열 변수를 숫자형으로 변환
mydata = pd.get_dummies(mydata, columns=['gender'], drop_first=True)
mydata
```

```python
mydata_corr = mydata.corr()
mydata_corr['married'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(mydata_corr, vmax=1, vmin=-1, center=0, annot=False, cmap='coolwarm', square=True)
#sns.heatmap(mydata_corr, vmax=1, vmin=-1, center=0, annot=True, cmap='coolwarm', fmt='.1f', square=True)
```
