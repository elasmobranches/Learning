```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# 경고 무시
warnings.filterwarnings("ignore")

# Data 경로 설정
DATA_DIR = "data"
```

```python
# 학습에 사용할 과거 주택매매 data set 로드 (dataset.csv)
dataset = pd.read_csv(os.path.join(DATA_DIR, "dataset.csv"))

# problem set 로드 (problem.csv)
problemset = pd.read_csv(os.path.join(DATA_DIR, "problem.csv"))
```

```python
# # 결측치가 있다면 GarageType은 NO Garage로, GarageYrBlt의 값은 0으로 대체
dataset["GarageType"].fillna("No Garage", inplace=True)
dataset["GarageYrBlt"].fillna(0, inplace=True)

dataset.isnull().sum()
```

```python
# dataset을 train set과 val set으로 나누기
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(dataset.drop("SalePrice", axis=1), dataset["SalePrice"], test_size=0.2, random_state=42)
```

```python
# 범주형 변수들을 모두 one-hot encoding하는데 x_train에 하는걸 그대로 x_val에도 적용
# x_train = pd.get_dummies(x_train)
# x_val = pd.get_dummies(x_val)
```

```python
# Random over-sampling 적용
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_train, y_train = ros.fit_resample(x_train, y_train)

```

오버샘플링을 하려고 했으나 소수 클래스에 대해 적은 수의 샘플이 있어서 SMOTE 또는 ADASYN이 적절한 이웃을 찾지 못하고 있다
그럼에도 불구하고 오버샘플링은 중요해보인다. 그렇다면 개수가 적은 샘플들은 다른 값으로 치환해 버릴까?

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# 수치형 변수와 범주형 변수를 구분
numeric_features = x_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = x_train.select_dtypes(include=['object']).columns

# 전처리 파이프라인 정의
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Pipeline 정의
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# train 데이터셋에 Pipeline 적용
x_train_preprocessed = pipeline.fit_transform(x_train)

# val 및 test 데이터셋에 Pipeline 적용
x_val_preprocessed = pipeline.transform(x_val)
x_problem_preprocessed = pipeline.transform(problemset)
```

```python
# lightgbm 모델 학습
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
lgb = lgb.LGBMRegressor()
lgb.fit(x_train_preprocessed, y_train)

```

```python
# 랜덤그리드 서치 이용
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint
from lightgbm import LGBMRegressor

# 하이퍼파라미터 그리드 정의
param_grid = {
    'n_estimators': sp_randint(30, 200),  # 트리의 개수
    'max_depth': sp_randint(3, 15),  # 트리의 최대 깊이
    'learning_rate': [0.01, 0.05, 0.1, 0.3, 0.5],  # 학습률
    'num_leaves': sp_randint(10, 100),  # 각 트리의 최대 잎의 수
    'subsample': [0.5, 0.7, 0.9, 1.0],  # 데이터 샘플링 비율
    'colsample_bytree': [0.5, 0.7, 0.9, 1.0]  # 열 샘플링 비율
}

# RandomizedSearchCV를 사용하여 랜덤 그리드 서치 수행
random_search = RandomizedSearchCV(estimator=LGBMRegressor(),
                                   param_distributions=param_grid,
                                   n_iter=500,  # 탐색할 조합의 수
                                   scoring='neg_mean_squared_error',  # 평가 지표
                                   cv=5,  # 교차 검증의 폴드 수
                                   verbose=1,
                                   random_state=42,
                                   n_jobs=-1)  # 병렬 처리 수


# 랜덤 그리드 서치를 사용하여 최적의 하이퍼파라미터 탐색
random_search.fit(x_train_preprocessed, y_train)

# 최적의 하이퍼파라미터 출력
print("Best hyperparameters:", random_search.best_params_)

# 최적의 모델로 예측 및 평가
best_model = random_search.best_estimator_
y_pred = best_model.predict(x_val_preprocessed)
print("Validation RMSE:", mean_squared_error(y_val, y_pred) ** 0.5)
```

```python
# 최적의 모델로 예측 및 평가
best_model = random_search.best_estimator_

# 최적의 모델로 훈련 데이터에 대한 예측
train_pred = best_model.predict(x_train_preprocessed)

# 최적의 모델로 검증 데이터에 대한 예측
test_pred = best_model.predict(x_val_preprocessed)

# 훈련 및 검증 데이터에 대한 RMSE 출력
print("Train RMSE :", mean_squared_error(y_train, train_pred) ** 0.5)
print("Val RMSE :", mean_squared_error(y_val, test_pred) ** 0.5)
```

```python
problem_pred = best_model.predict(x_problem_preprocessed)
```
