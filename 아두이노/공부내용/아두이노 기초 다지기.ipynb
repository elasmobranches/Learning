# 아두이노 기초 다지기

### 아두이노는 입력(센서), 출력(제어)을 할 수 있는 마이컴을 갖춘 소형 컴퓨터

다양한 스위치나 센서로부터 값을 받아들여 LED나 모터와 같은 전자 장치들로 출력을 제어함으로써 환경과 상호작용이 가능한 물건을 만들어 낼 수 있다.

### 아두이노 보드는 USB 케이블을 이용해서 컴퓨터와 통신 및 전원을 공급받는다

### 아두이노 보드에 따라서 맞는 드라이버와 시리얼 보드를 선택해야 한다

리셋버튼, 전원버튼(GND), 디지털 입출력, USB커넥터, 상태표시LED, DC전원 커넥터, 전원, 아날로그입력

디지털 입출력 - 외부로 0V, 5V값을 보내거나 외부에서 0V, 5V 값을 받는 역할을 한다 외부로 보낼 경우에는 LED등을 연결해 켜거나 끄는 동작을 수행하며 외부로부터 값을 받을 경우엔 버튼 등을 연결하여 버튼을 누르거나 떼는 동작을 판별하게 된다

아날로그 입력 핀은 외부로부터 아날로그 입력 값을 읽는 핀으로 주로 센서와 연결하여 사용된다. 아날로그 입력 값은 0~5V 사이의 전압 값을 256단계로 구분하여 읽게 된다

브래브보드  - 부품을 납땜하지 않고도 회로를 쉽게 구성할 수 있게 해주는 전기 회로 개발 도구. 내부적으로 연결되어 있어 전자 부품을 연결하기 편리하다

void -

함수의 반환 자료형으로 사용할 경우 "이 함수는 자료를 반환하지 않음."의 뜻을 가지게 된다.

void setup() { //~~ }

void loop() { //~}

;기호는 한 줄이 완료되었다는 표시이며 C언어의 기본 문법이다

‘//’는 프로그램에 대한 설명 즉 주석이다

setup함수는 코드를 실행할 때 한 번만 수행되며 사용하고자 하는 하드웨어를 초기화시키는 부분이다 모든 함수는 ‘{’로 시작해 ‘}’로 끝난다 

loop함수는 반복적으로 수행되며 하드웨어를 반복적으로 동작시키는 부분이다

아두이노 보드에 전원이 켜지면 위의 setup함수가 한 번 실행하고 난 뒤 loop 함수가 계속 실행된다. setup 함수 안에는 핀 설정 등과 같은 것이 들어가고 loop 함수에는 실제 어떤 작동을 일으키는 함수가 들어간다

```python
void setup() {
	Serial.begin(115200); # 아두이노와 통신할 때의 속도 - 시리얼모니터설정값과 동일해야
//'' 말고 ""써야한다
// C언어는 문장의 끝을 항상 세미콜론으로 끝낸다
	Serial.println("Hello")
```

컴파일은 인간이 읽을 수 있는 소스코드를 컴퓨터(기계)가 이해할 수 있는 혹은 실행할 수 있는 언어로 변경하는 과정(오류가 있는지 확인하는 과정이다)
